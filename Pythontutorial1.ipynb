```python
class Phone:
    def set_color(self,color):
        self.color = color
    def set_cost(self,cost):
        self.cost = cost
    def show_color(self):
        return self.color
    def show_cost(self):
        return self.cost
    def make_call(self):
        print("Make a call")
    def play_game(self):
        print("Play a game")
```

```python
class Employee:
    def __init__(self,name,age,salary,gender):
        self.name = name
        self.age = age
        self.salary = salary
        self.gender = gender
    
    def show_employee_details(self):
        print("Name of the employee",self.name)
        print("Age of the employee",self.age)
        print("Salary of the employee",self.salary)
        print("Gender of the employee",self.gender)
```

```python
e1 = Employee("Ram",23,25000,"Male")
```

```python
e1.show_employee_details()
```

```python
class Vehicle:
    def __init__(self,mileage,cost):
        self.mileage = mileage
        self.cost = cost
        
    def show_vehicle_details(self):
        print("Mileage of the vehicle is",self.mileage)
        print("Cost of the vehivle is",self.cost)
        print("I am vehicle")
```

```python
v1 = Vehicle(20,450)
```

```python
v1.show_vehicle_details()
```

```python
class Car(Vehicle):
    def show_car_details(self):
        print("I am a car")
```

```python
c1 = Car(250,500)
```

```python
c1.show_vehicle_details()
```

```python
c1.show_car_details()
```

```python
class Car(Vehicle):
    def __init__(self,mileage,cost,tyres,hp):
        super().__init__(mileage,cost)
        self.tyres = tyres
        self.hp = hp
        
    def show_car_details(self):
        print("Number of tyres in a car",self.tyres)
        print("Horse power in a car",self.hp)
        print("I am a car")
```

```python
c1 = Car(600,1000000000,8,999)
```

```python
c1.show_car_details()
```

```python
c1.show_vehicle_details()
```
